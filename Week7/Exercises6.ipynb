```python
a = -9.8 #constant acceleration of gravity
t1 = 0
t2 = 3
dt = 0.1
t = arange(t1,t2,dt)
v = np.zeros(t.shape)
r = np.zeros(t.shape)
v[0] = 0 #initial velocity = 0 m/s
r[0] = 15#initial height = 15 meters
```

```python
for i in range(1, len(t)):
    v[i] = v[i-1] + a*dt
    r[i] = r[i-1] + v[i-1]*dt
```

```python
plot(t,v,label='velocity',linewidth=2)
plot(t,r,label='position',linewidth=2,color='red')
title('Free Fall',fontsize = 20)
xlabel('time (s)', fontsize = 15)
ylabel('height (m)', fontsize = 15)
legend()
show()
```

This is the expected graph for free fall data.  The velocity decreases linearly because the acceleration of gravity is constant, while the position decreases parabolically as the object falls faster and faster.

##### Euler-Cromer Method

```python
v2 = np.zeros(t.shape)
r2 = np.zeros(t.shape)
v2[0] = 0 
r2[0] = 15 
for i in range(1,len(t)):
    v2[i] = v2[i-1] + a*dt
    r2[i] = r2[i-1] + v2[i]*dt #using just calculated velocity instead of previous velocity
```

```python
plot(t,v2,linewidth = 2,label ='velocity')
plot(t,r2,color = 'red', linewidth = 2, label = 'position')
title('Free Fall using the Euler-Cromer Method',fontsize=18)
xlabel('time (s)',fontsize=15)
ylabel('height (m)',fontsize=15)
legend()
show()
```

This gives results similiar to the previous method so I know it must be working.  The position seems to have changed a bit, possibly giving a more accurate answer.

##### Midpoint Method

```python
v3 = np.zeros(t.shape)
r3 = np.zeros(t.shape)
v3[0] = 0 
r3[0] = 15 
for i in range(1,len(t)):
    v3[i] = v3[i-1] + a*dt
    r3[i] = r3[i-1] + (1/2.)*(v3[i]+v3[i-1])*dt
```

```python
plot(t,v3,linewidth = 2,label ='velocity')
plot(t,r3,color = 'red', linewidth = 2, label = 'position')
title('Free Fall using the Midpoint Method',fontsize=18)
xlabel('time (s)',fontsize=15)
ylabel('height (m)',fontsize=15)
legend()
show()
```

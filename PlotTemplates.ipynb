#### Line Plots

```python
y = arange(-5,5.1,0.1) #Creating range for the x-axis
E = cosh(y) #Defining the functions I want to graph
p = sinh(y)
```

```python
plot(y,E,color="blue",linewidth=2,label="cosh(y)") #plotting, setting the color, size, and label of the data
plot(y,p,color="orange",linewidth=2,label="sinh(y)")
grid(True)
title("Generic Plot Template",fontsize = 20)
xlabel("x-axis (s)",fontsize =14) #setting the axis labels
ylabel("E(y),|p|(y)",fontsize = 14)
ax = gca()
ax.spines['right'].set_color('none') #erasing the bounding box from the plot
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
legend(loc = 'lower right') #moving the legend so it doesn't cover the data
show() #creating the plot without extra words also being shown
```

#### Line Plots

```python
y = arange(-5,5.1,0.1) #Creating range for the x-axis
E = cosh(y) #Defining the functions I want to graph
p = sinh(y)
```

```python
plot(y,E,color="blue",linewidth=2,label="cosh(y)") #plotting, setting the color, size, and label of the data
plot(y,p,color="orange",linewidth=2,label="sinh(y)")
grid(True)
title("Generic Plot Template",fontsize = 20)
xlabel("x-axis (s)",fontsize =14) #setting the axis labels
ylabel("E(y),|p|(y)",fontsize = 14)
ax = gca()
ax.spines['right'].set_color('none') #erasing the bounding box from the plot
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
legend(loc = 'lower right') #moving the legend so it doesn't cover the data
show() #creating the plot without extra words also being shown
```

#### Subplots

```python
y = arange(-5,5.1,0.1) #Creating range for the x-axis
E = cosh(y) #Defining the functions I want to graph
p = sinh(y)
subplot(211) #This means a 2 row, 1 column subplot and I'm plotting in the 1st plot
plot(y,E,color="blue",linewidth=2,label="cosh(y)")
subplot(212) #This means the same 2 row, 1 column subplot but now I'm plotting in the 2nd plot
plot(y,p,color="orange",linewidth=2,label="sinh(y)")
xlabel("x-axis")

show()

```

#### Scatter Plots

```python
n = 200 #getting random data for scatter plot
x = random.normal(0,1,n)
y = random.normal(0,1,n)
```

```python
scatter(x,y,color="purple",marker='s') #creating plot of x and y and changing the style and color of the points
xlabel("x-axis",fontsize=14)
ylabel("y-axis",fontsize=14)
title("Sample Scatter Plot",fontsize=20)
show()
```

Electric Dipole Equation
-

\begin{equation}
V(x,y) = \frac{kq}{\sqrt{x^2 + (y^2-d)}} - \frac{kq}{\sqrt{x^2 + (y^2+d)}}
\end{equation}

```python
dx = 0.06
dy = 0.06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
print x.ndim
print len(x)
```

```python
X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
```

```python
print Y
```

```python
#make the function V(x,y)
k=1; q=1; d=1
V = (k*q/(X**2+(Y-d)**2)**(1/2.)) - (k*q/(X**2+(Y+d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.autumn)
cb = fig.colorbar(p, shrink=0.5) #have to assign it to the plot 'p' and then make it smaller with shrink
```

```python
#Represent 3D data with contours
levels = array([8,4,2,1,0.5,-0.25,-0.5,-1,-2,-4,-8]) #Define the levels for the contours so they aren't evenly spaced
contour(X,Y,V,levels,linewidths=2,cmap=cm.autumn)
show()
```

Ex 2: Determining Electric Potential
-

```python
dx = 8./101.
dy = 8./101.
x = arange(-4.,4.,dx)
y = arange(-4.,4.,dy)
X,Y = meshgrid(x,y)
print Y.shape
```

```python
k=8.987551787e9 #units of k: N*m^2/C^2
q=1e-9 #units of q: C
d=2 #units of d: m
V = (k*q/(X**2+(Y-d)**2)**(1/2.)) + (k*q/(X**2+(Y+d)**2)**(1/2.)) 
U = -(k*q/((X-d)**2+Y**2)**(1/2.)) - (k*q/((X+d)**2+Y**2)**(1/2.))
```

```python
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V+U, rstride=1, cstride=1, linewidth=0, cmap=cm.autumn)
cb = fig.colorbar(p, shrink=0.5)
title("Potential Distribution of a Quadpole (V)",fontsize=16)
```

```python
levels = array([8,4,2,1,0.5,-0.25,-0.5,-1,-2,-4,-8]) #Define the levels for the contours so they aren't evenly spaced
contour(X,Y,V+U,levels,linewidths=2,cmap=cm.autumn)
title("Contour Map of Quadpole",fontsize=16)
show()
```

Ex 3: Functions
-

Part a

```python
def pointPotential(x,y,q,posx,posy):
    """This function will take an x-y coord, charge, and position of the charge and return the potential at that x-y point"""
    k=8.987551787e9 #units of k: N*m^2/C^2
    Vxy = (k*q/((x-posx)**2+(y-posy)**2)**(1/2.))
    return Vxy
```

```python
x=1
y=1
V11 = pointPotential(x,y,1e-9,0.,0.)
print V11
```

```python
x,y = meshgrid(arange(-2.,2.,0.06),arange(-2.,2.,0.06))
Vxy = pointPotential(x,y,1e-9,0.,0.)
print Vxy
```

Part b

```python
def dipolePotential(x,y,q,d):
    """This function will determine the potential at x,y due to a dipole, instead of a single point charge"""
    Vxy = pointPotential(x,y,q,-d,0.) - pointPotential(x,y,q,d,0.)
    return Vxy
```

```python
x=1;y=1
V11 = dipolePotential(x,y,1e-9,1)
print V11
```

Part c

Dipole:

```python
x,y = meshgrid(arange(-2.,2.,0.06),arange(-2.,2.,0.06))
Vxy = dipolePotential(x,y,1e-9,1)
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.ocean_r)
cb = fig.colorbar(p, shrink=0.5)
title("Dipole graph using functions",fontsize=16)
```

Quadpole:

```python
dx = 8./101.
dy = 8./101.
x,y = meshgrid(arange(-4.,4.,dx),arange(-4.,4.,dy))
q = 1e-9 #in Coloumbs
d = 2 #in meters
Vxy = pointPotential(x,y,-q,d,0.) + pointPotential(x,y,-q,-d,0.)+ pointPotential(x,y,q,0.,d) + pointPotential(x,y,q,0.,-d)
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.ocean_r)
cb = fig.colorbar(p, shrink=0.5)
title("Quadpole graph using pointPotential functions",fontsize=16)
```

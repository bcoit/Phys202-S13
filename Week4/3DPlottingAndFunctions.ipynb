Electric Dipole Equation
-

\begin{equation}
V(x,y) = \frac{kq}{\sqrt{x^2 + (y^2-d)}} - \frac{kq}{\sqrt{x^2 + (y^2+d)}}
\end{equation}

```python
dx = 0.06
dy = 0.06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
print x.ndim
print len(x)
```

```python
X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
```

```python
print Y
```

```python
#make the function V(x,y)
k=1; q=1; d=1
V = (k*q/(X**2+(Y-d)**2)**(1/2.)) - (k*q/(X**2+(Y+d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.autumn)
cb = fig.colorbar(p, shrink=0.5) #have to assign it to the plot 'p' and then make it smaller with shrink
```

```python
#Represent 3D data with contours
levels = array([8,4,2,1,0.5,-0.25,-0.5,-1,-2,-4,-8]) #Define the levels for the contours so they aren't evenly spaced
contour(X,Y,V,levels,linewidths=2,cmap=cm.autumn)
show()
```

$G(x,y) = e^{-x^{2}-y^{2}}

```python
dx = 12./51.
dy = 12./51.
x = arange(-6.,6.,dx)
y = arange(-6.,6.,dy)
print len(x)
```

```python
X,Y = meshgrid(x,y)
print Y.shape
```

```python
G = e**(-X**2-Y**2)
```

```python
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, G, rstride=1, cstride=1, linewidth=0, cmap=cm.Greens_r)
cb = fig.colorbar(p, shrink=0.5)
cset = ax.contour(X,Y,G,array([0.,0.2,0.4,0.6,0.8]),zdir='x',offset=-6.,cmap=cm.Greens)
```

Ex 2: Determining Electric Potential
-

```python
dx = 8./101.
dy = 8./101.
x = arange(-4.,4.,dx)
y = arange(-4.,4.,dy)
X,Y = meshgrid(x,y)
print Y.shape
```

```python
k=8.987551787*10**9; q=1e**(-9); d=2 #Look up k up to 8 sig figs and then put a comment in with units for all the values K:N*m^2/C^2
V = (k*q/(X**2+(Y-d)**2)**(1/2.)) + (k*q/(X**2+(Y+d)**2)**(1/2.)) 
U = -(k*q/((X-d)**2+Y**2)**(1/2.)) - (k*q/((X+d)**2+Y**2)**(1/2.))
```

```python
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V+U, rstride=1, cstride=1, linewidth=0, cmap=cm.autumn)
cb = fig.colorbar(p, shrink=0.5)
```

```python
t = arange(-4.5,4.51,0.03)
f1 = (t+1)/(t-1)
f2 = e**(-t)
```

```python
plot(t,f1,label='(t+1)/(t-1)')
plot(t,f2,color='red',linestyle='dotted',label='exp(-t)')
ylim(-10,10)
xlim(-4.3,4.3)
legend(loc='lower left')
xlabel('time (seconds)',fontsize=14)
ylabel('f(t), g(t)',fontsize = 14)
annotate("discontinuous", (1.1,-2), fontsize = 14) 
annotate("here", (1.15, -3.5), fontsize = 14)
show()
```

```python
import scipy.integrate as integrate
```

```python
a = 1 #m
b = 2 #m
charge_density = 1.0e-6 #C/m
x0 = 3; y0 = 3 #m
k = 8.987551787e9 #N*m^2/C^2
```

```python
dEx = lambda x: k*charge_density*(x0-x)/(((x0-x)**2 + (y-y0)**2)**(1/2.))
```

```python
dEy = lambda y: k*charge_density*(y0-y)/(((y0-y)**2 + (x-x0)**2)**(1/2.))
```

```python
integrate.quad(dEx,a,b)
```

```python
integrate.quad(dEy,a,b)
```

##### Part b

```python
def is_anagram(s1, s2):
    """Takes in two words as aruguments and returns True if they are anagrams of each other and False if they are not"""
    if len(s1) != len(s2):
        return False
    else:
        for i in range(0,len(s1)+1):
            if s1[i] in s2:
                test = True
            else:
                test = False
                return test
            return test
```

```python
is_anagram('frog','grof') #Testing true arguments
```

```python
is_anagram('grump','grumpy') #Testing arguments of 2 different lengths
```

```python
is_anagram('happy','large') #Testing False arguments of the same length
```

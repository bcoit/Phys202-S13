```python
t = arange(0.0,1.01,0.01)
c = cos(4*pi*t)
s = sin(4*pi*t)
```

```python
c
```

```python
plot(t,c,color="blue",linewidth=2.5,label="$cos(4{\pi}t)$")
plot(t,s,color="red",linewidth=2.5,label="$sin(4{\pi}t)$")
xlabel("voltage (mV)")
ylabel("time (s)")
title("My Cool Graph")
grid(True)
legend()
savefig("simple_plot.pdf")
show()
```

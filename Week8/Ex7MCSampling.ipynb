Example 1: Analytic Inverse CDF Transform
-

```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
    
plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
N = 10000
z = numpy.random.random_sample(N);
Finv = lambda z: arccos(cos(0)-2*z)
xsamples = Finv(z);
hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,f(x),label="True distribution",lw=3,color='red')
xlim(0,pi)
ylim(0,0.71)
legend(loc=0)
show()
```

Exercise 1: Exponential function - analytic
-

$f(x)=\lambda e^{-\lambda x}$ from 0 to $\infty$ and 0 otherwise

$F(x) = -e^{-\lambda x}+1$

$F^{-1}=-\frac{log(1-x)}{\lambda}$

```python
l = 1
x = arange(0,5.01,0.1)
f = l*e**(-x*l)
F = -e**(-x*l)+1
plot(x,f,label='True distribution',lw=3)
plot(x,F,label='CDF',lw=3)
legend(loc='center right')
show()
```

```python
N = 10000
z = numpy.random.random_sample(N);
Finv = lambda z: -log(1-z)/l
xsamples = Finv(z);
hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,f,label="True distribution",lw=3,color='red')
legend(loc=0)
xlim(0,5)
show()
```

```python
N = 10000
z = numpy.random.random_sample(N)
```

```python
def distribute1D(x,prob,N):
    """takes any numerical distribution prob, on the interval defined by the array x, and returns
an array of N sampled values that are statistically the same as the input data."""
    F = cumsum(prob)/(sum(prob))
    z = numpy.random.random_sample(N)
    xsamples = x[searchsorted(F,z)]
    return xsamples
```

Hw Problem #2
-

$F(x)=1-e^{-(\frac{x}{\beta})^{\alpha}}$

$F^{-1}=-\beta ln(1-x)^{-\alpha}$

```python

N = 10000
b = 6
a = 1.5
x = arange(0,16.01,0.1)
f = lambda x: a*b**(-a)*x**(a-1)*e**(-(x/b)**a)
CDF = lambda x: 1-e**(-(x/b)**a)
z = numpy.random.random_sample(N);
Finv = lambda z: b*(-log(z))**(1/a)
xsamples = Finv(z);
hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,CDF(x),label="CDF",lw=3,color='blue')
plot(x,f(x),label="True distribution",lw=3,color='red')
legend(loc='center right')
xlim(0,16)
title('Weibull distribution',fontsize = 18)
show()
```
